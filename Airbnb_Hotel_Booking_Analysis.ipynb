```python
import pandas as pd

def clean_airbnb_data(file_path):
    df = pd.read_csv(file_path)
    df.drop_duplicates(inplace=True)
    df.drop(['house_rules', 'license'], axis=1, inplace=True, errors='ignore')
    if 'price' in df.columns:
        df['price'] = df['price'].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False)
    if 'service fee' in df.columns:
        df['service fee'] = df['service fee'].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False)
    df.rename(columns={
        'price': 'price_$',
        'service fee': 'service_fee_$'
    }, inplace=True)
    df.dropna(inplace=True)
    if 'price_$' in df.columns:
        df['price_$'] = pd.to_numeric(df['price_$'], errors='coerce')
    if 'service_fee_$' in df.columns:
        df['service_fee_$'] = pd.to_numeric(df['service_fee_$'], errors='coerce')
    if 'id' in df.columns:
        df['id'] = df['id'].astype(str)
    if 'host id' in df.columns:
        df['host id'] = df['host id'].astype(str)
    if 'last review' in df.columns:
        df['last review'] = pd.to_datetime(df['last review'], errors='coerce')
    if 'Construction year' in df.columns:
        df['Construction year'] = pd.to_numeric(df['Construction year'], errors='coerce').astype('Int64')

    # Correct spelling mistake: 'brookln' -> 'Brooklyn'
    if 'neighbourhood group' in df.columns:
        df.loc[df['neighbourhood group'] == 'brookln', 'neighbourhood group'] = 'Brooklyn'

    # Remove outliers in availability_365 (values > 500 are invalid)
    if 'availability 365' in df.columns:
        df = df.drop(df[df['availability 365'] > 500].index)

    # Drop rows that became NaN after conversions
    df.dropna(inplace=True)

    return df
```

```python
hood_group = df.groupby('neighbourhood group').size().reset_index(name='count')
hood_group_bar = plt.bar(hood_group['neighbourhood group'], hood_group['count'])
plt.bar_label(hood_group_bar, labels=hood_group['count'], padding=4)

# Formatting
plt.ylim([0, 50000])
plt.xlabel('Neighborhood Groups')
plt.ylabel('Number of Listings')
plt.xticks(rotation=45)
plt.title('Which Neighborhood Group has the Highest Number of Listings?')

plt.show()
```

```python
avg_price = df.groupby('neighbourhood group')['price'].mean().sort_values(ascending=False).to_frame()
avg_price_bar = plt.bar(avg_price.index, avg_price['price'], color='skyblue')
plt.bar_label(avg_price_bar, labels=round(avg_price['price'], 2), label_type='edge', padding=4)
plt.ylim([0, 700])
plt.xlabel('Neighborhood Group')
plt.ylabel('Average Price per Listings ($)')
plt.title('Average Price per Listings ($) in each Neighborhood Group')
plt.xticks(rotation=45)
plt.show()
```

```python
avg_price_by_year = df.groupby('Construction year')['price'].mean().to_frame()
avg_price_by_year.plot(figsize=(10,5), color='green', marker='o')
plt.xlabel('Construction Year')
plt.ylabel('Average Price ($)')
plt.title('Average Price ($) for Properties in each Construction Year')
plt.grid(True)
plt.show()
```

```python
hosts = df.groupby('host name')['calculated host listings count'].sum().nlargest(7).to_frame()
hosts_bar = plt.bar(hosts.index, hosts['calculated host listings count'], color='orange')
plt.bar_label(hosts_bar, labels=hosts['calculated host listings count'], label_type='edge', padding=3)
plt.xlabel('Host Name')
plt.ylabel('Calculated Host Listings Count')
plt.title('Top 10 Hosts by Calculated Host Listings Count')

# Formatting
plt.xticks(rotation=80)
plt.ylim([0, 180000])
plt.show()
```
